### II.I. Preliminary Data Cleaning and Reformatting

With the intention of adapting the information that was provided to be
able to replicate the analysis, a series of initial transformations are
carried out. Throughout this subsection, a new data set will be created
that is a cleaned and suitable version for analysis. The new data set is
called *Titanic_clean*, and the modifications carried out are:

-   Elimination of empty rows. It was identified that the original data
    set contained rows that were completely empty. They are promptly
    dropped.

-   Blanks. It was identified that many cells contained blanks, they
    were replaced by *NA*.

-   Survived reformatting. For ease of understanding and to make the
    variable categorical rather than numerical, the *survived* variable
    is modified in the following way:

    -   An original value of *0* is transformed to "*No*", to clearly
        represent that the passenger did not survive.

    -   An original value of *1* is transformed to "*Yes*", to clearly
        represent that the passenger survived.

-   Passenger class reformatting. To make the variable categorical
    rather than numerical, the *pclass* variable is modified in the
    following way:

    -   An original value of *1* is transformed to "*1st Class*".

    -   An original value of *2* is transformed to "*2nd Class*".

    -   An original value of *3* is transformed to "*3rd Class*".

-   Passenger sex reformatting. This is an aesthetic modification,
    simply adding an uppercase to the beginning of the variable.

-   Port of Embarkation reformatting. For ease of understanding, the
    *embarked* variable is modified in the following way:

    -   An original value of *S* is transformed to "*Southampton*".

    -   An original value of *Q* is transformed to "*Queenstown*".

    -   An original value of *C* is transformed to "*Cherburg*".

Any variable that does not fit with our expectations is transformed to
*NA*.


````r
# Data Cleaning
Titanic_clean <- Titanic.Data

Titanic_clean <- Titanic_clean %>% filter(Reduce(`+`, lapply(., is.na)) != ncol(.))

Titanic_clean[Titanic_clean==''] <- NA

Titanic_clean$survived <- ifelse(Titanic_clean$survived == 0 , "No" ,
                          ifelse(Titanic_clean$survived == 1 , "Yes" , NA))

Titanic_clean$pclass <- ifelse(Titanic_clean$pclass == 1 , "1st Class" ,
                        ifelse(Titanic_clean$pclass == 2 , "2nd Class" ,
                        ifelse(Titanic_clean$pclass == 3 , "3rd Class" , NA)))

Titanic_clean$sex <- ifelse(Titanic_clean$sex == "female" , "Female" ,
                     ifelse(Titanic_clean$sex == "male" , "Male" , NA))

Titanic_clean$embarked <- ifelse(Titanic_clean$embarked == "S" , "Southampton" ,
                          ifelse(Titanic_clean$embarked == "Q" , "Queenstown" ,
                          ifelse(Titanic_clean$embarked == "C" , "Cherburg" , NA)))
````

````r
# Main variables
Total_Passengers <- nrow(Titanic_clean)

Total_Survived <- length(which(Titanic_clean$survived == "Yes"))
Total_Died <- length(which(Titanic_clean$survived == "No"))
Survival_Rate = Total_Survived / Total_Passengers

Total_Class_1 <- length(which(Titanic_clean$pclass == "1st Class"))
Total_Class_2 <- length(which(Titanic_clean$pclass == "2nd Class"))
Total_Class_3 <- length(which(Titanic_clean$pclass == "3rd Class"))
Class_1_Rate = Total_Class_1 / Total_Passengers
Class_2_Rate = Total_Class_2 / Total_Passengers
Class_3_Rate = Total_Class_3 / Total_Passengers

Total_Female <- length(which(Titanic_clean$sex == "Female"))
Total_Male <- length(which(Titanic_clean$sex == "Male"))
Female_Rate <- Total_Female / Total_Passengers
Male_Rate <- Total_Male / Total_Passengers

Total_Cherburg <- length(which(Titanic_clean$embarked == "Cherburg"))
Total_Southampton <- length(which(Titanic_clean$embarked == "Southampton"))
Total_Queenstown <- length(which(Titanic_clean$embarked == "Queenstown"))
Cherburg_Rate = Total_Cherburg / Total_Passengers
Southampton_Rate = Total_Southampton / Total_Passengers
Queenstown_Rate = Total_Queenstown / Total_Passengers

Mean_Age <- mean(subset(Titanic_clean , is.na(Titanic_clean$age) == FALSE)$age)
Mean_Fare <- mean(subset(Titanic_clean , is.na(Titanic_clean$fare) == FALSE)$fare)

# Outputs
print(paste("The total number passengers is:", Total_Passengers))
print(paste("The number passengers that survived is:", Total_Survived))
print(paste("The number passengers that perished is:", Total_Died))
print(paste("The survival rate is: ", round(100*Survival_Rate, 2), "%" , sep=""))
print(paste("The average age is:", round(Mean_Age, 2) , "years"))
print(paste("The average fare is: $", round(Mean_Fare, 2) , sep=""))
```

### II.III. Exploratory Analysis

In this subsection the following exploratory analysis are carried out:

1.  Analysis of the passenger sex variable

2.  Analysis of the port of embarkation variable

3.  Analysis of the age variable

4.  Analysis of the passenger class variable

5.  Analysis of the survival variable

6.  Correlation and comments

#### II.III.I Passenger Sex

It can be seen that there are more males than females. Also, that
proportionally, 1st class has more females than males and that 3rd class
has more males than females. Finally, passenger sex seems to be a good
predictor of the survival rate as shown in Figure 3.

```{r echo= FALSE , message = FALSE , warning = FALSE}
# Figure 1: Port of Embarkation by Sex
Grouped_Figure_1 <- Titanic_clean %>% group_by(embarked , sex) %>% summarize(Number_of_passangers = n())

f1_plot1 <- ggplot(data = Grouped_Figure_1 , aes( x = sex , y = Number_of_passangers , fill = embarked ))+
  scale_fill_manual(values=c(Cherburg, Queenstown , Southampton))+
  xlab("Sex")+
  ylab("Number of Passengers")+
  ggtitle("" , subtitle = "Port of Embarkation by Sex")+
  guides(fill = guide_legend(title = "Port of Embarkation"))+
  geom_bar(position="stack", stat="identity")

f1_plot2 <- ggplot(data = Grouped_Figure_1 , aes( x = sex , y = Number_of_passangers , fill = embarked ))+
  scale_fill_manual(values=c(Cherburg, Queenstown , Southampton))+
  xlab("Sex")+
  ylab("Percentage")+
  scale_y_continuous(labels = scales::label_percent(accuracy = 1))+
  ggtitle("" , subtitle = "Port of Embarkation Rate by Sex")+
  guides(fill = guide_legend(title = "Port of Embarkation"))+
  geom_bar(position="fill", stat="identity")+
  geom_hline(aes(yintercept = Southampton_Rate) , colour = "darkgray" , linetype = "dashed" , size = 1)+
  geom_hline(aes(yintercept = Southampton_Rate + Queenstown_Rate) , colour = "darkgray" , linetype = "dashed" , size = 1)

Figure1 <- ggarrange(f1_plot1, f1_plot2 , common.legend = TRUE, legend = "bottom", ncol = 2)
Figure1 <- annotate_figure(Figure1, fig.lab = "Figure 1: Port of Embarkation by Sex" , fig.lab.face = "bold" , fig.lab.size = 12)

Figure1
````

(unique sex, class and port of embankment) is assessed and plotted in
the flowing charts. It is evident that 1st class and female passengers

### II.IV. Final Data Preparation

Following the description of the variables, the quality assessment, and
the exploratory analysis, the following is carried out:

-   Dropping of variables. The following variables are dropped name,
    cabin, boat, body, home.dest.

-   Passenger class is transformed into 2 dummy variables, is_class1 and
    is_class2 The variable is then dropped.

-   Ticket is transformed into is_alone, completed with "Yes" and "No"
    depending on the case. The original variable is then dropped.

-   Embarked is transformed into 2 dummy variables, is_southampton and
    is_queenstown. The variable is then dropped. Also, the 2 women of
    1st class that don't have a port of embarkation are dropped.

-   All passengers who did not pay a fare (fare of 0) are dropped.

````r
# Dropping of flagged observations
Titanic_clean <- Titanic_clean[!is.na(Titanic_clean$embarked),]
Titanic_clean <- subset(Titanic_clean , Titanic_clean$fare != 0)

# Dropping of variables
Titanic_dummy <- Titanic_clean[ , -which(names(Titanic_clean) %in% c("name","cabin","boat","body","home.dest"))]

# Transforming Passenger Class
Titanic_dummy$is_class1 <- ifelse(Titanic_dummy$pclass == "1st Class","Yes","No")
Titanic_dummy$is_class2 <- ifelse(Titanic_dummy$pclass == "2nd Class","Yes","No")
Titanic_dummy <- Titanic_dummy[ , -which(names(Titanic_dummy) %in% c("pclass"))]

# Transforming Ticket
Ticket_count <- Titanic_dummy %>% group_by(ticket) %>% summarize(passenger_number = n())
Ticket_count$is_alone <- ifelse(Ticket_count$passenger_number == 1 , "Yes" , "No")
list_of_alone <- subset(Ticket_count , Ticket_count$is_alone == "Yes")$ticket
Titanic_dummy$is_alone <- ifelse(Titanic_dummy$ticket %in% list_of_alone , "Yes" , "No")
Titanic_dummy <- Titanic_dummy[ , -which(names(Titanic_dummy) %in% c("ticket"))]

# Transforming embarked
Titanic_dummy$is_southampton <- ifelse(Titanic_dummy$embarked == "Southampton","Yes","No")
Titanic_dummy$is_queenstown <- ifelse(Titanic_dummy$embarked == "Queenstown","Yes","No")
Titanic_dummy <- Titanic_dummy[ , -which(names(Titanic_dummy) %in% c("embarked"))]

# Output
print(paste("The cleaned dataset, called Titanic_clean has", nrow(Titanic_clean) , "observations and" , length(Titanic_clean) , "variables"))
print(paste("The final dataset that incorporates the dummy variables and droppes the unnecesary ones, called Titanic_dummy has", nrow(Titanic_dummy) , "observations and" , length(Titanic_dummy) , "variables"))
````

````r
str(Titanic_dummy)
head(Titanic_dummy)
summary(Titanic_dummy)
````

````r
# Count total NA values
total_na <- sum(is.na(Titanic_dummy))
cat("Total NA values in the dataset:", total_na, "\n")
# Count NA values per column
na_per_column <- sapply(Titanic_dummy, function(x) sum(is.na(x)))
print(na_per_column)
````

````r
# Remove rows with any NA values
Titanic_dummy_clean <- na.omit(Titanic_dummy)
cat("Number of observations after removing NAs:", nrow(Titanic_clean), "\n")
# Convert character columns to factors
char_cols <- sapply(Titanic_dummy_clean, is.character)
Titanic_dummy_clean[char_cols] <- lapply(Titanic_dummy_clean[char_cols], as.factor)
str(Titanic_dummy_clean)
````

````r
# Feature Scaling
numeric_cols <- c("age", "sibsp", "parch", "fare")
preProcValues <- preProcess(Titanic_dummy_clean[, numeric_cols], method = c("center", "scale"))
Titanic_dummy_clean[, numeric_cols] <- predict(preProcValues, Titanic_dummy_clean[, numeric_cols])
````

````r
# Split Data
set.seed(1023)
train_index <- createDataPartition(Titanic_dummy_clean$survived, p = 0.7, list = FALSE)
training_data <- Titanic_dummy_clean[train_index, ]
testing_data <- Titanic_dummy_clean[-train_index, ]
cat("Training set size:", nrow(training_data), "\n")
cat("Testing set size:", nrow(testing_data), "\n")
````

````r
# Training Control
train_control <- trainControl(
  method = "cv",
  number = 10,
  classProbs = TRUE,
  summaryFunction = twoClassSummary,
  savePredictions = TRUE
)
````

````r
# Train Models
# Logistic regression
lr_model <- train(
  survived ~ .,
  data = training_data,
  method = "glm",
  family = "binomial",
  trControl = train_control,
  metric = "ROC"
)
# Random forest
rf_model <- train(
  survived ~ .,
  data = training_data,
  method = "rf",
  trControl = train_control,
  metric = "ROC",
  tuneLength = 5
)
# SVM
svm_model <- train(
  survived ~ .,
  data = training_data,
  method = "svmRadial",
  trControl = train_control,
  metric = "ROC",
  tuneLength = 5
)
# XGBoost
xgb_model <- train(
  survived ~ .,
  data = training_data,
  method = "xgbTree",
  trControl = train_control,
  metric = "ROC",
  tuneLength = 2
)
````

````r
# Compare Models
models_list <- list(
  Logistic_Regression = lr_model,
  Random_Forest = rf_model,
  SVM = svm_model,
  XGBoost = xgb_model)
# Consolidate resampling results
model_results <- resamples(models_list)
summary(model_results)
# Select Best Model
best_model_name <- names(which.max(colMeans(model_results$values[, grep("ROC", colnames(model_results$values))])))
cat("Best model based on ROC AUC:", best_model_name, "\n")
best_model <- models_list[[best_model_name]]
````

````r
# View all hyperparameter combinations and their performance for xgb_model
print(xgb_model$results)
# View the parameters for the final XGBoost model
final_xgboost <- xgb_model$finalModel
print(final_xgboost$params)
````
